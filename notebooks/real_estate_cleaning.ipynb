# Data Cleaning

```python
df.columns # view all column names
```

```python
df.head()
```

### Converting the column names to lower case

```python
df.columns = df.columns.str.lower() # converting column names to lower case
```

```python
df.head()
```

### Dropping the null values

```python
df.dropna(inplace=True) #dropping the null values and replacing the df
df.isnull().sum()[df.isnull().sum() > 0] #checking if any null values present after dropping
```

### Checking for duplicates and removing them

```python
df[df.duplicated()]
```

### Correcting errors

```python
# correcting spelling mistakes in area, sale_cond,park_facil,buildtype,utility,street,mzzone

print(sorted(df['area'].unique()))
```

```python
df['area'] = df['area'].replace({'Adyr':'Adyar',
                    'Ann Nagar': 'Anna Nagar',
                    'Ana Nagar': 'Anna Nagar',
                   'Chormpet':'Chrompet',
                   'Chrompt':'Chrompet',
                    'Chrmpet': 'Chrompet',
                   'KKNagar':'KK Nagar',
                   'Karapakam':'Karapakkam',
                   'TNagar': 'T Nagar',
                   'Velchery': 'Velachery'})
```

```python
sorted(df['sale_cond'].unique())
```

```python
df['sale_cond'] = df['sale_cond'].replace({'Ab Normal': 'Abnormal',
                        'AbNormal':'Abnormal',
                        'AdjLand':'Adj Land',
                        'PartiaLl':'Partial',
                        'Partiall':'Partial'})
```

```python
df['park_facil'] = df['park_facil'].replace('Noo', 'No')
```

```python
df['buildtype'].unique()
```

```python
df['buildtype'] = df['buildtype'].replace({'Comercial': 'Commercial',
                        'Other': 'Others'
                        })
```

```python
df['utility_avail'].unique()
```

```python
df['utility_avail'] = df['utility_avail'].replace('AllPub', 'All Pub')
```

```python
df['street'].unique()
```

```python
df['street'] = df['street'].replace({'Pavd':'Paved',
                     'NoAccess':'No Access'})
```

# Data Tranformation

```python
df.columns 
```

```python
# converting data type for date columns
df.date_build = pd.to_datetime(df.date_build, format='%d-%m-%Y') 
df.date_sale = pd.to_datetime(df.date_sale, format='%d-%m-%Y')
```

```python
df.dtypes #to view the data tyoes of all columns for making necessary transformations
```

```python
# conveting N-bedroom and n_bathroom column to integer data type
df.n_bathroom = df.n_bathroom.astype(int)
df.n_bedroom = df.n_bedroom.astype(int)

df.info()
```

# Data Aggregation

```python
# creating new columns - property age, price per sqft, total price

df['prop_age'] = df.date_sale.dt.year - df.date_build.dt.year
df.head()
```

```python
df['price_per_sqft'] = df.sales_price / df.int_sqft
```

```python
df['price_per_sqft'] = df.price_per_sqft.astype(float).round(2)
```

```python
df['total_price'] = df.sales_price + df.commis + df.reg_fee
```

```python
df['qs_avg'] = ((df.qs_rooms + df.qs_bathroom + df.qs_bedroom) / 3).round(2)
```
